```python
%matplotlib inline
import matplotlib.pyplot as plt
import random
plt.style.use('ggplot')

# Helper function to compute cumulative sum of a list of numbers
def cum_sum(l):
    s = 0.0
    for item in l:
        s += item
        yield s
  
# Generate something to plot
r = range(1000)
gwn = [random.normalvariate(0.0, 1.0) for x in r]
random_walk = list(cum_sum(gwn))

# Plot it
plt.plot(r, random_walk)
```
